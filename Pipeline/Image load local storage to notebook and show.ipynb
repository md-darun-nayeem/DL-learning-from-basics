```python
import os
```

```python
from PIL import Image
```

```python
root = 'images/cat'
```

```python
fnames = os.listdir(root)
```

```python
len(fnames)
```

```python
# file names

for i in range(10):
    print(fnames[i])
```

```python
import matplotlib.pyplot as plt
```

```python
fig,axs = plt.subplots(nrows=2,ncols=5,figsize=(7,7))
axs = axs.flatten()

for i in range(10):
    file_path = os.path.join(root,fnames[i])
    img = Image.open(file_path)
    axs[i].imshow(img)
    axs[i].axis('off')
    #axs[i].set_title(fnames[i])
plt.show()
```
